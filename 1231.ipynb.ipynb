```python
# 2.2 data.txt 내용을 헥사코드로 변환 후 hex_data.txt에 저장
```

```python
FILE_NAME=r'D:\EXAM_python\data.txt'
with open(FILE_NAME,mode='r',encoding='utf-8') as r:
    text=r.read()

FILE_NAME=r'D:\EXAM_python\hex_data.txt'
with open(FILE_NAME,mode='w',encoding='utf-8') as w:
    for i in text:
        w.write(i if i =='\n' else hex(ord(i)))

with open(FILE_NAME,mode='r',encoding='utf-8') as r:
    print(r.read())
```

```python
# 3)
# 파일명 : data.txt
# 기능 : 3줄 화면 출력 후 + 입력 시 다음 3줄 출력
#       파일 끝까지 반복하여 모두 출력
```

```python
FILE_NAME=r'D:\EXAM_python\data.txt'
fp=open(FILE_NAME, mode='r',encoding='utf-8')
line=fp.readlines()
fp.close()
i=0
while True:
    print(*line[i:i+3])
    입력=input('+ 입력시 3줄 더 출력 : ')
    if 입력=='+': i=i+3
    else: break
    if i==9: i=0
```

```python
# 4) 2번문제 2개의 파일을 한 줄씩 합쳐서 새로운 all_data.txt 파일 생성
```

```python
FILE_NAME1=r'D:\EXAM_python\data.txt'
FILE_NAME2=r'D:\EXAM_python\hex_data.txt'
FILE_NAME3=r'D:\EXAM_python\all_data.txt'
with open(FILE_NAME1, mode='r',encoding='utf-8') as r1:
    read1=r1.readlines()
    with open(FILE_NAME2, mode='r',encoding='utf-8') as r2:
        read2=r2.readlines()
        with open(FILE_NAME3, mode='w',encoding='utf-8') as w:
            for i in range(len(read1)):
                w.write(read1[i]+read2[i])
with open(FILE_NAME3,mode='r',encoding='utf-8') as f:
    print(f.read())
```
